```python
# for displaying tables
import pandas as pd
from IPython.display import display

# for database
import sqlite3

conn = sqlite3.connect("repos.db")
c = conn.cursor()
```

summary statistics

```python
c.execute("SELECT count(*) FROM repos")
rows = c.fetchall()
num_rows = rows[0][0]
print("Number of Collected Repos: %d" % num_rows)
```

```python
stars_threshold = 5
c.execute("SELECT count(*) FROM repos WHERE stars >= %d" % stars_threshold)
rows = c.fetchall()
num_rows = rows[0][0]
print("Number of Collected Repos with More than %d Stars: %d" % (stars_threshold, num_rows))
```

display top results with most number of stars that have not been checked over

```python
result_size = 30
query = "SELECT id, url, stars FROM repos WHERE checked=0 ORDER BY stars DESC LIMIT %d" % result_size
df = pd.read_sql_query(query, conn)
display(df)
```

```python
c.close()
conn.close()
```
